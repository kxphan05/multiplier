```python
import requests
import re


all_links = []
for page_no in range(1, 61):
    url = "https://grail.moe/library?keyword=economics&doc_type=Notes%2FPractices&page={page_no}#google_vignette"

    # This is the "Base" pattern we know exists
    api_pattern = "https://api.grail.moe/note/download/"

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
    }

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()
        html_text = response.text
        print(html_text)

        # 1. Broad Search: Look for the base URL followed by 1 or more digits
        # The [0-9]+ part ensures we grab the ID number
        links = re.findall(r'\\"id\\":[0-9]+', html_text)
        links = [f"{match.split(':')[1]}" for match in links]

        # Clean up results
        unique_links = filter(lambda x: int(x) > 10000, sorted(list(set(links)), reverse=True))
        unique_links = [link for link in unique_links]
        all_links.extend(unique_links)

    except Exception as e:
        print(f"Connection error: {e}")
```
